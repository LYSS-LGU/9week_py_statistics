## Ïù¥ÏÇ∞ ÌôïÎ•† Î∂ÑÌè¨ (Discrete Probability Distribution)

ÌôïÎ•† Î≥ÄÏàòÍ∞Ä Ï∑®Ìï† Ïàò ÏûàÎäî Í∞íÏù¥ Ïú†ÌïúÌïòÍ±∞ÎÇò ÏÖÄ Ïàò ÏûàÎäî Í≤ΩÏö∞, Ïù¥ÏÇ∞ÌôïÎ•†Î≥ÄÏàò XÏùò ÌôïÎ•†Î∂ÑÌè¨

- Ïù¥ÏÇ∞ÌôïÎ•†Î≥ÄÏàòÎäî 0, 1, 2, 3, ... Í≥º Í∞ôÏù¥ Í∞úÎ≥ÑÏ†ÅÏúºÎ°ú ÏÖÄ Ïàò ÏûàÎäî Í∞íÏùÑ Í∞ñÎäîÎã§.
- Ïù¥Ìï≠Î∂ÑÌè¨, Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨, Ìè¨ÏïÑÏÜ° Î∂ÑÌè¨ Îì±
- ÌôïÎ•†ÏßàÎüâÌï®Ïàò(Probability Mass Function: PMF)Î°ú ÌëúÌòÑ
- Í∞Å Ïù¥ÏÇ∞Ï†ÅÏù∏ ÌôïÎ•†Í∞íÏù¥ ÏñºÎßàÎÇò ÎêòÎäîÏßÄÎ•º Ìï®ÏàòÎ°ú ÌëúÌòÑÌïòÏó¨ Í∞Å ÌôïÎ•†Î≥ÄÏàò Í∞íÏùò ÌôïÎ•†ÏùÑ ÏßÅÏ†ëÏ†ÅÏúºÎ°ú Í≥ÑÏÇ∞
- Ïù¥ÏÇ∞ÌôïÎ•†Î≥ÄÏàòÏùò PMFÎäî P(X=x)Î°ú ÌëúÍ∏∞ÌïòÎ©∞, Î™®Îì† Í∞ÄÎä•Ìïú Í∞í xÏóê ÎåÄÌï¥ ÌôïÎ•†Ïùò Ìï©ÏùÄ 1Ïù¥ Îê©ÎãàÎã§.

Ïù¥ÏÇ∞Ìòï ÌôïÎ•†Î≥ÄÏàòÏùò ÏòàÏãú: ÎèôÏ†Ñ 2Í∞úÎ•º ÎçòÏ†∏ ÏïûÎ©¥/Îí∑Î©¥Ïù¥ ÎÇòÏò§Îäî Í≤ΩÏö∞Ïùò Ïàò(H: Ïïû, T: Îí§)
| ÏÇ¨Í±¥ | HH | HT | TH | TT | Ìï©Í≥Ñ |
| --- | --- | --- | --- | --- | --- |
| P(X) | 1/4 | 1/4 | 1/4 | 1/4 | 1 |

```python
# üìÅ ÏúÑÏπò: dice_prob_plot.py

import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd

# ‚úÖ ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  # ÎòêÎäî 'AppleGothic' (Mac), 'NanumGothic' (Linux)

# ‚úÖ ÎßàÏù¥ÎÑàÏä§ Íπ®Ïßê Î∞©ÏßÄ
matplotlib.rcParams['axes.unicode_minus'] = False

# Ï£ºÏÇ¨ÏúÑ Îàà (1~6)
x = np.arange(1, 7)
probs = [1/6] * 6

# ÏãúÍ∞ÅÌôî
plt.bar(x, probs)
plt.xticks(x)
plt.xlabel("Ï£ºÏÇ¨ÏúÑ Îàà")
plt.ylabel("ÌôïÎ•†")
plt.title("Ï£ºÏÇ¨ÏúÑ ÌôïÎ•†Î∂ÑÌè¨")
plt.grid(axis='y')
plt.show()
```

### Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨ (Bernoulli Distribution)

ÌôïÎ•† Î≥ÄÏàò XÍ∞Ä ÏÑ±Í≥µ ÎòêÎäî Ïã§Ìå®ÎùºÎäî 2Í∞ÄÏßÄ Í≤∞Í≥º Ï§ë ÌïòÎÇòÎßå Í∞ñÎäî ÌôïÎ•†Î∂ÑÌè¨. 

ÏÑ±Í≥µÌôïÎ•†Í≥º Ïã§Ìå®ÌôïÎ•†Ïùò Ìï©ÏùÄ 1Ïù¥Î©∞, Í∞Å ÏãúÌñâÏùÄ ÏÑúÎ°ú ÎèÖÎ¶ΩÏ†ÅÏûÖÎãàÎã§.

### ÏòàÏãú)  ÎèôÏ†ÑÏùÑ ÎçòÏ†∏ ÏïûÎ©¥Ïù¥ÎÇò Îí∑Î©¥Ïù¥ ÎÇòÏò§Îäî ÏÇ¨Í±¥

ÏïÑÎûò ÏΩîÎìúÎäî Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Î•º PythonÏúºÎ°ú Íµ¨ÌòÑÌïú ÏòàÏãúÏûÖÎãàÎã§. p=0.5Ïù∏ ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞Î•º ÏãúÎÆ¨Î†àÏù¥ÏÖòÌïòÍ≥†, 1,000Î≤à ÎçòÏ°åÏùÑ ÎïåÏùò Í≤∞Í≥ºÎ•º Î∂ÑÏÑùÌïòÏó¨ Ïù¥Î°†Ï†Å ÌôïÎ•†Í∞íÍ≥º Ïã§ÌóòÍ∞íÏùÑ ÎπÑÍµêÌï©ÎãàÎã§.

```python
# Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨ ÏòàÏãú - ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨ ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
p = 0.5  # ÏïûÎ©¥Ïù¥ ÎÇòÏò¨ ÌôïÎ•†

# Î≤†Î•¥ÎàÑÏù¥ ÌôïÎ•†Î≥ÄÏàò ÏÉùÏÑ±
X = bernoulli(p)

# 1Ìöå ÏãúÌñâ Í≤∞Í≥º ÏãúÎÆ¨Î†àÏù¥ÏÖò (0: Îí∑Î©¥, 1: ÏïûÎ©¥)
result = X.rvs(size=1)
print(f"ÎèôÏ†Ñ 1Ìöå ÎçòÏßÄÍ∏∞ Í≤∞Í≥º: {'ÏïûÎ©¥' if result[0] == 1 else 'Îí∑Î©¥'}")

# Ïó¨Îü¨ Î≤à ÏãúÌñâ Í≤∞Í≥º ÏãúÎÆ¨Î†àÏù¥ÏÖò
n_trials = 1000
results = X.rvs(size=n_trials)

# Í≤∞Í≥º Î∂ÑÏÑù
success_count = np.sum(results)
success_rate = success_count / n_trials

print(f"1000Î≤à ÎçòÏ°åÏùÑ Îïå ÏïûÎ©¥ ÎÇòÏò® ÌöüÏàò: {success_count}")
print(f"ÏïûÎ©¥ ÎÇòÏò¨ ÌôïÎ•† (Ïù¥Î°†Í∞í: {p}, Ïã§ÌóòÍ∞í: {success_rate:.4f})")

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
plt.bar([0, 1], [1-p, p], color=['blue', 'red'])
plt.xticks([0, 1], ['Îí∑Î©¥ (0)', 'ÏïûÎ©¥ (1)'])
plt.ylabel('ÌôïÎ•†')
plt.title('Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨ (p=0.5Ïù∏ ÎèôÏ†Ñ ÎçòÏßÄÍ∏∞)')
plt.show()
```

### Ïù¥Ìï≠ Î∂ÑÌè¨ (Binomial Distribution)

ÏùºÏ†ïÌïú ÌöüÏàòÏùò ÎèÖÎ¶ΩÏ†ÅÏù∏ Î≤†Î•¥ÎàÑÏù¥ ÏãúÌñâÏóêÏÑú **ÏÑ±Í≥µÏùò ÌöüÏàòÎ•º ÎÇòÌÉÄÎÇ¥Îäî Î∂ÑÌè¨**

‚Üí nÎ≤àÏùò Î≤†Î•¥ÎàÑÏù¥ ÏãúÌñâÏóê kÎ≤à ÏÑ±Í≥µÌï† ÌôïÎ•†Ïù¥ pÏù∏ ÌôïÎ•† Î∂ÑÌè¨

Ïù¥Ìï≠Î∂ÑÌè¨Ïùò ÏãúÌñâ ÌöüÏàòÍ∞Ä ÎßéÏïÑÏßÄÎ©¥ Ï†ïÍ∑úÎ∂ÑÌè¨ÏôÄ Ïú†ÏÇ¨Ìï¥Ïßê.

- ÌôïÎ•†Î≥ÄÏàò ùëã : ùëõÎ≤àÏùò ÎèÖÎ¶ΩÏ†ÅÏù∏ ÏãúÌñâ Ï§ë ÏÑ±Í≥µ ÌöüÏàò
- ÌôïÎ•†ÏßàÎüâÌï®Ïàò (PMF) : ùëòÎ≤à Î∞úÏÉùÌï† ÌôïÎ•†

![Ïù¥Ìï≠Î∂ÑÌè¨](images\Ïù¥Ìï≠Î∂ÑÌè¨.png)
![Ïù¥Ìï≠Î∂ÑÌè¨2](images\Ïù¥Ìï≠Î∂ÑÌè¨2.png)

Ïù¥Ìï≠Î∂ÑÌè¨Îäî Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Î•º nÌöå/ nÎ≤àÏù¥ ÏàòÌñâÌï©ÎãàÎã§.  

Ï¶â Î≤†Î•¥ÎàÑÏù¥ Î∂ÑÌè¨Ïóê ÌöüÏàòÍ∞Ä Ï∂îÍ∞ÄÎêòÏñ¥ Í≥±Ìï¥ÏßÑÎã§ > ÌöüÏàòÍ∞Ä ÎßéÏïÑÏßÄÎ©¥ Ï†ïÍ∑úÎ∂ÑÌè¨ÏôÄ Ïú†ÏÇ¨ÌïòÎã§.

### ÏòàÏãú : ÎèôÏ†Ñ Ïó¨Îü¨ Î≤à ÎçòÏßÄÍ∏∞ - ÎèôÏ†ÑÏùÑ 10Î≤à ÎçòÏ†∏ÏÑú ÏïûÎ©¥Ïù¥ ÎÇòÏò§Îäî ÌöüÏàò.

ÏïÑÎûò ÏΩîÎìúÎäî Ïù¥Ìï≠ Î∂ÑÌè¨ B(10, 0.5)Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÎèôÏ†ÑÏùÑ 10Î≤à ÎçòÏ°åÏùÑ Îïå ÏïûÎ©¥Ïù¥ ÎÇòÏò§Îäî ÌöüÏàòÏóê ÎåÄÌïú ÌôïÎ•†Î∂ÑÌè¨Î•º Í≥ÑÏÇ∞ÌïòÍ≥† ÏãúÍ∞ÅÌôîÌï©ÎãàÎã§.
- Ïù¥Ìï≠Î∂ÑÌè¨Ïùò ÌôïÎ•†ÏßàÎüâÌï®Ïàò(PMF)ÏôÄ ÎàÑÏ†ÅÎ∂ÑÌè¨Ìï®Ïàò(CDF)Î•º Í∑∏ÎûòÌîÑÎ°ú ÌëúÌòÑÌïòÏó¨ Í∞Å Í≤ΩÏö∞Ïùò ÌôïÎ•†ÏùÑ ÏâΩÍ≤å ÌååÏïÖÌï† Ïàò ÏûàÏäµÎãàÎã§

```python
# Ïù¥Ìï≠ Î∂ÑÌè¨ ÏòàÏãú - ÎèôÏ†Ñ 10Î≤à ÎçòÏßÄÍ∏∞
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Ïù¥Ìï≠ Î∂ÑÌè¨ ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
n = 10  # ÏãúÌñâ ÌöüÏàò
p = 0.5  # ÏÑ±Í≥µ ÌôïÎ•†(ÎèôÏ†Ñ ÏïûÎ©¥Ïù¥ ÎÇòÏò¨ ÌôïÎ•†)

# Ïù¥Ìï≠ ÌôïÎ•†Î≥ÄÏàò ÏÉùÏÑ±
X = binom(n, p)

# Í∞ÄÎä•Ìïú Î™®Îì† Í≤∞Í≥ºÍ∞í(0Î∂ÄÌÑ∞ nÍπåÏßÄ)
k_values = np.arange(0, n+1)

# Í∞Å Í≤∞Í≥ºÍ∞íÏùò ÌôïÎ•† Í≥ÑÏÇ∞
pmf_values = X.pmf(k_values)

# Í≤∞Í≥º ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values)
plt.xlabel('ÏïûÎ©¥Ïù¥ ÎÇòÏò§Îäî ÌöüÏàò (k)')
plt.ylabel('ÌôïÎ•† P(X=k)')
plt.title(f'Ïù¥Ìï≠ Î∂ÑÌè¨ B({n}, {p}) - ÎèôÏ†Ñ 10Î≤à ÎçòÏßÄÍ∏∞')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# ÌèâÍ∑†Í≥º Î∂ÑÏÇ∞ Ï∂úÎ†•
mean = n * p
var = n * p * (1-p)
plt.figtext(0.7, 0.8, f'ÌèâÍ∑†: {mean}', fontsize=12)
plt.figtext(0.7, 0.75, f'Î∂ÑÏÇ∞: {var}', fontsize=12)

plt.show()

# ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò(CDF) Í≥ÑÏÇ∞ Î∞è ÏãúÍ∞ÅÌôî
cdf_values = X.cdf(k_values)

plt.figure(figsize=(10, 6))
plt.step(k_values, cdf_values, where='post', color='red')
plt.scatter(k_values, cdf_values, color='red')
plt.xlabel('ÏïûÎ©¥Ïù¥ ÎÇòÏò§Îäî ÌöüÏàò (k)')
plt.ylabel('ÎàÑÏ†Å ÌôïÎ•† P(X‚â§k)')
plt.title(f'Ïù¥Ìï≠ Î∂ÑÌè¨ B({n}, {p})Ïùò ÎàÑÏ†Å Î∂ÑÌè¨ Ìï®Ïàò')
plt.xticks(k_values)
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

### Ìë∏ÏïÑÏÜ° Î∂ÑÌè¨ (Poisson Distribution)

Ï£ºÏñ¥ÏßÑ ÏãúÍ∞ÑÍ≥º Í≥µÍ∞Ñ ÎÇ¥ÏóêÏÑú Ïñ¥Îñ§ ÏÇ¨Í±¥Ïùò Î∞úÏÉù ÌöüÏàòÏóê ÎåÄÌïú ÌôïÎ•†Î∂ÑÌè¨ÏûÖÎãàÎã§. 

Ïù¥ Î∂ÑÌè¨Îäî Í∏∞ÎåÄÍ∞íÍ≥º Î∂ÑÏÇ∞Ïù¥ Î™®Îëê ÎûåÎã§ Í∞íÏúºÎ°ú ÎèôÏùºÌïòÎ©∞, ÎèÖÎ¶ΩÏÑ±, ÎπÑÎ°ÄÏÑ±, ÎπÑÏßëÎùΩÏÑ±Ïùò ÌäπÏßïÏùÑ Í∞ÄÏßÄÍ≥† ÏûàÏùå

- Îã®ÏúÑ ÏãúÍ∞Ñ(ÎòêÎäî Îã®ÏúÑ Í≥µÍ∞Ñ)ÏóêÏÑú ÌäπÏ†ï ÏÇ¨Í±¥Ïùò Î∞úÏÉù ÎπàÎèÑÎ•º ÎÇòÌÉÄÎÇ¥Îäî Î∂ÑÌè¨
- ÎìúÎ¨ºÍ≤å Î∞úÏÉùÌïòÎäî ÏÇ¨Í±¥Ïùò ÎπàÎèÑÎ•º Î™®Îç∏ÎßÅÌï† Í≤ΩÏö∞ ÏÇ¨Ïö©Îê®
- ÌôïÎ•†Î≥ÄÏàò ùëã : ÏùºÏ†ïÌïú ÏãúÍ∞ÑÏóêÏÑú Î∞úÏÉùÌïòÎäî ÏÇ¨Í±¥Ïùò ÌöüÏàò
- ÌôïÎ•†ÏßàÎüâÌï®Ïàò (PMF) : (ùúÜÎäîÎã®ÏúÑ ÏãúÍ∞ÑÏóêÏÑú ÏÇ¨Í±¥Ïù¥ Î∞úÏÉùÌïòÎäî ÌèâÍ∑† ÌöüÏàò)

![Ìë∏ÏïÑÏÜ°Î∂ÑÌè¨](images\Ìë∏ÏïÑÏÜ°Î∂ÑÌè¨.png)

### 1ÏãúÍ∞Ñ ÎÇ¥Ïóê Î≤ÑÏä§Í∞Ä ÎèÑÏ∞©ÌïòÎäî ÌöüÏàòÏóê ÎåÄÌïú ÌôïÎ•† Î∂ÑÌè¨ Í≥ÑÏÇ∞

```python
# Ìë∏ÏïÑÏÜ° Î∂ÑÌè¨ - Î≤ÑÏä§ ÎèÑÏ∞© ÏãúÎÆ¨Î†àÏù¥ÏÖò
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# ÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
lambda_val = 3  # 1ÏãúÍ∞ÑÎãπ ÌèâÍ∑† Î≤ÑÏä§ ÎèÑÏ∞© ÌöüÏàò

# Ìë∏ÏïÑÏÜ° ÌôïÎ•†Î≥ÄÏàò ÏÉùÏÑ±
X = poisson(lambda_val)

# Í∞ÄÎä•Ìïú Î≤ÑÏä§ ÎèÑÏ∞© ÌöüÏàò (0Î∂ÄÌÑ∞ 10ÍπåÏßÄ)
k_values = np.arange(0, 11)

# Í∞Å Í∞íÏóê ÎåÄÌïú ÌôïÎ•† Í≥ÑÏÇ∞
pmf_values = X.pmf(k_values)

# Í≤∞Í≥º ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values)
plt.xlabel('1ÏãúÍ∞Ñ ÎÇ¥ Î≤ÑÏä§ ÎèÑÏ∞© ÌöüÏàò')
plt.ylabel('ÌôïÎ•†')
plt.title(f'Ìë∏ÏïÑÏÜ° Î∂ÑÌè¨ (Œª = {lambda_val})')
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# ÌèâÍ∑†Í≥º Î∂ÑÏÇ∞ ÌëúÏãú
plt.annotate(f'ÌèâÍ∑†(Œª) = {lambda_val}\nÎ∂ÑÏÇ∞ = {lambda_val}', 
             xy=(0.7, 0.8), xycoords='axes fraction',
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

plt.show()

```

```python
# ÏãúÎÆ¨Î†àÏù¥ÏÖò: ÌïòÎ£®(24ÏãúÍ∞Ñ) ÎèôÏïàÏùò Î≤ÑÏä§ ÎèÑÏ∞© Ìå®ÌÑ¥
np.random.seed(42)
hours = np.arange(1, 25)
arrivals = np.random.poisson(lambda_val, size=24)

plt.figure(figsize=(12, 6))
plt.bar(hours, arrivals)
plt.xlabel('ÏãúÍ∞ÑÎåÄ')
plt.ylabel('Î≤ÑÏä§ ÎèÑÏ∞© ÌöüÏàò')
plt.title('Ìë∏ÏïÑÏÜ° Î∂ÑÌè¨Î•º Ïù¥Ïö©Ìïú 24ÏãúÍ∞Ñ Î≤ÑÏä§ ÎèÑÏ∞© ÏãúÎÆ¨Î†àÏù¥ÏÖò')
plt.xticks(hours)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Ïó∞ÏÜç ÌôïÎ•† Î∂ÑÌè¨ (Continuous Probability Distribution)

ÌôïÎ•† Î≥ÄÏàòÍ∞Ä Ï∑®Ìï† Ïàò ÏûàÎäî Í∞íÏù¥ Ïó∞ÏÜçÏ†ÅÏù∏ Í≤ΩÏö∞(Ïã§ÏàòÏùò Ïñ¥Îäê ÌäπÏ†ï Íµ¨Í∞Ñ Ï†ÑÏ≤¥)Ïóê Ìï¥ÎãπÌïòÎäî ÌôïÎ•† Î≥ÄÏàò

# Ï†ïÍ∑ú Î∂ÑÌè¨ (Normal Distribution)

- ÌèâÍ∑†Ïóê Í∞ÄÏû• ÎßéÏù¥ Î™∞Î†§ ÏûàÍ≥† ÌèâÍ∑†Ïóê Î©ÄÏñ¥ÏßàÏàòÎ°ù ÌôïÎ•†(ÎπàÎèÑ)Í∞Ä ÎÇÆÏïÑÏßÑÎã§.
- Î∂ÑÏÇ∞Ïù¥ ÌÅ¥ÏàòÎ°ù ÏñëÏòÜÏúºÎ°ú ÌçºÏßÄÎäî Î™®Ïñë,Íº¨Î¶¨Î∂ÄÎ∂ÑÏù¥ ÎëêÍ∫ºÏõåÏßÑÎã§.

### ÏòàÏãú) 
- Ïñ¥Îäê ÏãúÌóòÏóêÏÑú Ïö∞Îì±Î∞òÏóê Îì§Ïñ¥Í∏∞Í∞Ä ÏúÑÌï¥ÏÑúÎäî ÏÉÅÏúÑ 2% ÏïàÏóê Îì§Ïñ¥Ïïº ÌïúÎã§.
- Ìï¥Îãπ ÏãúÌóò Ï†êÏàòÎäî Ï†ïÍ∑úÎ∂ÑÌè¨Î•º Îî∞Î•¥Î©∞, ÌèâÍ∑†Ïù¥ 85Ï†êÏù¥Í≥† ÌëúÏ§ÄÌé∏Ï∞®Í∞Ä 5Ïùº Îïå, Ïö∞Îì±Î∞òÏóê Îì§Ïñ¥Í∏∞Í∞Ä ÏúÑÌïú ÏµúÏÜå ÏãúÌóòÏ†êÏàòÎäî?  (Îã® P(Z ‚â§ 2. 05) = 0. 98)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Ï†ïÍ∑úÎ∂ÑÌè¨ ÏòàÏãú ÏΩîÎìú
mu = 85  # ÌèâÍ∑†
sigma = 5  # ÌëúÏ§ÄÌé∏Ï∞®

# Ï†ïÍ∑úÎ∂ÑÌè¨Ïùò ÌôïÎ•†Î∞ÄÎèÑÌï®Ïàò
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
pdf = stats.norm.pdf(x, mu, sigma)

# ÏÉÅÏúÑ 2%Ïóê Ìï¥ÎãπÌïòÎäî Ï†êÏàò Í≥ÑÏÇ∞
z_score = 2.05
critical_value = mu + z_score * sigma

# ÏãúÍ∞ÅÌôî
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, 'b-', lw=2, label='Ï†ïÍ∑úÎ∂ÑÌè¨ PDF')
plt.axvline(critical_value, color='r', linestyle='--', label=f'ÏÉÅÏúÑ 2% Í≤ΩÍ≥Ñ: {critical_value:.2f}')
plt.fill_between(x, pdf, where=(x >= critical_value), alpha=0.3, color='r')

# Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
plt.title('ÏãúÌóò Ï†êÏàòÏùò Ï†ïÍ∑úÎ∂ÑÌè¨')
plt.xlabel('Ï†êÏàò')
plt.ylabel('ÌôïÎ•†Î∞ÄÎèÑ')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
